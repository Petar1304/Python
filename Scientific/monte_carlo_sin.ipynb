```python
import numpy as np
import matplotlib.pyplot as plt

N = 100000

X = []
Y = []
sinX = []
sinY = []

for _ in range(N):
    x = np.random.uniform(0, 2*np.pi)
    y = np.random.uniform(-1, 1)
    
    if y >= 0:
        if y <= np.sin(x):
            sinX.append(x)
            sinY.append(y)
        else:
            X.append(x)
            Y.append(y)
            
    if y < 0:
        if y >= np.sin(x):
            sinX.append(x)
            sinY.append(y)
        else:
            X.append(x)
            Y.append(y)
        
        
# plotting
plt.figure(1)
plt.plot(X, Y, 'g.')
plt.plot(sinX, sinY, 'r.')
plt.show()

totalArea = 4 * np.pi 
area = totalArea * (len(sinX)/N)
print(f'Total area: {totalArea}\nArea under the function: {area}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000

logX = []
logY = []
X = []
Y = []

for _ in range(N):
    x = np.random.uniform(0, 30)
    y = np.random.uniform(0, 1/np.e)
    
    if y < np.log(x)/x:
        logX.append(x)
        logY.append(y)
    else:
        X.append(x)
        Y.append(y)
        

    
# plot
plt.figure(1)
plt.plot(X, Y, 'y.')
plt.plot(logX, logY, 'b.')
plt.grid()
    
    
# calculate area
totalArea = 30 * 1/np.e
area = (len(logX) / N) * totalArea

print(f'Total area: {totalArea}\nArea under the function: {area}')
```

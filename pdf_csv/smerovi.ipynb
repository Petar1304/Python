```python
import csv

file = '/home/petar/Documents/pdf_csv/smerovi.csv'
```

```python
with open(file, 'r', encoding='UTF-8') as f:
    reader = csv.reader(f)    
    df = []
    for i in reader:
        data = {}
        if i[0] == '0':
            continue
        
        data['rang'] = i[0] 
        data['index'] = i[1]
        data['ime'] = '{} {}'.format(i[2], i[3])
        data['lista'] = i[-4]
        data['smer'] = i[-3]
        data['koeficijent'] = i[-1]
        df.append(data)
```

```python
df[0]
```

```python
of = []
ir = []
os = []
og = []
ot = []
oe = []

for i in df:
    if i['smer'] ==  'ОФ':
        of.append(i)
    
    elif i['smer'] ==  'ИР':
        ir.append(i)
    
    elif i['smer'] ==  'ОС':
        os.append(i)
    
    elif i['smer'] ==  'ОГ':
        og.append(i)
        
    elif i['smer'] ==  'ОТ':
        ot.append(i)
    
    elif i['smer'] ==  'ОЕ':
        oe.append(i)
```

```python
print('fizicka = {}\nrti = {}\nsignali = {}\nenergetika = {}\ntelekomunikacije = {}\nelektronika = {}'.format(len(of), len(ir),len(os), len(og), len(ot), len(oe)))
```
